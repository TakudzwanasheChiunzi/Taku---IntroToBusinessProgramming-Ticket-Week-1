# Step 5 – Write the Code
Translate your pseudocode into Python code. Below is one example solution:

````python
# TAKU'S ATTEMPT
# Import datetime to timestamp your work
from datetime import datetime

# Collect user information
username = input("Howdy stranger! What do they call you in these parts?")
theme = input("What app theme suits your fancy? (e.g., gold, anime, dark) ")
currency = input("What currency are you using? (e.g., BWP, MUR, ZAR) ")
goals = input("Now why are you trying to save money?...seriously why? (e.g., New Car, Holiday) ")
goals2 = input("Is there anything else your saving your money towards?")

# Create the welcome message
summary = ("Here you Go! Your very own Personal Finance Tracker, " + username + "! You're one step closer to being financially aware! " + \
           "As per your liking this app will use a " + theme + " theme and all amounts will be in " + \
           currency + ". Since you wanna save up for " + \
           goals + " we'll make sure to get all your ducks in a row! And don't you worry, we'll help you save up for " + goals2 + " as well!")

# Display the summary message
print(summary)
````
