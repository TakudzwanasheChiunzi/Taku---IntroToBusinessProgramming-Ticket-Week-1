## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### üí° Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""My Budget Tracker Prompt:Okay Gem, so my next coding task is to create a budget tracker. I need a simple, easy to understand by a beginner, python program/code that will: Allows users to input their monthly income, Allows users to input up to 3 expenses, Calculates remaining budget, Shows a warning if budget is negative. Please create this for me."""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# # 1. Initialize variables
total_expenses = 0.0
expense_count = 1

# 2. Get the monthly income
monthly_income = float(input("Enter your total monthly income: "))

# 3. Collect 3 expenses using a loop
print("\nPlease enter your top 3 monthly expenses:")
while expense_count <= 3:
    expense_value = float(input(f"Enter amount for expense {expense_count}: "))
    total_expenses = total_expenses + expense_value # Add to the running total
    expense_count = expense_count + 1               # Move to the next count

# 4. Calculate the remaining balance
remaining_budget = monthly_income - total_expenses

# 5. Display the results and the warning logic
print("-" * 30)
print(f"Total Income:   ${monthly_income:,.2f}")
print(f"Total Expenses: ${total_expenses:,.2f}")
print(f"Remaining:      ${remaining_budget:,.2f}")

# 6. The "Warning" Logic (Conditional Statement)
if remaining_budget < 0:
    print("\n‚ö†Ô∏è WARNING: You are over budget! Your expenses exceed your income.")
elif remaining_budget == 0:
    print("\nNote: You have exactly $0 left. Your budget is perfectly balanced.")
else:
    print("\nGreat job! You are within your budget.")
```
