## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []  # Store all expense amounts (as numbers)
descriptions = []  # Store what each expense was for (as strings)
total_spent = 0.0 # Initialize as float for calculations

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

# Your task: Complete the multi-expense entry system
# Use sentinel value pattern to collect multiple expenses
# Validate each expense amount (positive numbers only)
# Store both amount and description
# Show running total and budget percentage after each entry

while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    try:
        expense_amount = float(expense_input) # Convert input to a float
        if expense_amount <= 0:
            print("Expense must be a positive number!")
            continue # Skip to next iteration to ask for input again
    except ValueError:
        print("Invalid amount. Please enter a number.")
        continue # Skip to next iteration to ask for input again

    # Get the description and store it in a *separate* temporary variable
    current_description = input("Write what this expense was for: ")

    expenses.append(expense_amount) # Append the numeric expense amount
    descriptions.append(current_description) # Append the description string

    total_spent += expense_amount # Add the numeric expense amount to the total
    percentage_of_budget = (total_spent / monthly_budget) * 100

    print(f"  Current total spent: ${total_spent:.2f}")
    print(f"  Budget used: {percentage_of_budget:.1f}%")

print(f"\n📊 Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals they want to track
num_goals = 0
while num_goals < 1 or num_goals > 3:
    try:
        num_goals = int(input("How many goals do you have for this month? (1-3): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 3:
            print("Have only 3 main goals for this month (goals that are SMART and can be achieved by this month)")
    except ValueError:
        print("Please enter a valid number of goals!")

# Your task: Collect goal information using for loop
# For each goal (1 to num_goals):
#   - Get goal name
#   - Get target amount (with validation)
#   - Get current amount (with validation)
#   - Store in the three lists

print(f"\nEntering {num_goals} savings goals:")

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    #Your goal collection code here

    current_goal_name = input("What is the name of your goal?: ")
    goal_names.append(current_goal_name)

    current_goal_target = -1.0
    while current_goal_target < 0:
        try:
            goal_targets_input_str = input(f"How much are do you want to save for your goal {current_goal_name}? $")
            current_goal_target = float(goal_targets_input_str)
            if current_goal_target < 0:
                print("Your goal target must be a positive number!")
        except ValueError:
            print("Please enter a valid number for the target!")
    goal_targets.append(current_goal_target)

    current_goal_current = -1.0
    while current_goal_current < 0:
        try:
            goal_current_input_str = input(f"How much have you saved currently for {current_goal_name}? $")
            current_goal_current = float(goal_current_input_str)
            if current_goal_current < 0:
                print("Current savings cannot be negative!")
        except ValueError:
            print("Please enter a valid number for current savings!")
    goal_current.append(current_goal_current)


print("\n" + "="*30)
print("YOUR SAVINGS GOALS")
print("="*30)

for i in range(len(goal_names)):

    if goal_targets[i] > 0:
        percent = (goal_current[i] / goal_targets[i]) * 100
    else:
        percent = 0.0

    print(f"Goal: {goal_names[i]}")
    print(f"Progress: ${goal_current[i]:.2f} / ${goal_targets[i]:.2f} ({percent:.1f}%)")
    print("-" * 15)

print(f"\n✅ Saved {len(goal_names)} goals!")
```

```python
# Goal Analysis and Progress Tracking
if len(goal_names) > 0:
    print("\n📊 Goals Progress Analysis:")

    # Your task: Analyze each goal
    # For each goal, calculate:
    #   - Progress percentage
    #   - Amount remaining
    #   - Months to completion (if saving $100/month)

    total_target = 0
    total_current = 0

    for i in range(len(goal_names)):
        # Your progress calculation here

        total_target += goal_targets[i]
        total_current += goal_current[i]

    # Overall progress summary
    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\n🎯 Overall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")

    # Your task: Find which goal needs most attention
    # Find the goal with lowest progress percentage
    # Provide encouragement and specific advice

    lowest_percent = 101.0
    attention_goal = ""

    total_target_analysis = 0
    total_current_analysis = 0

    for i in range(len(goal_names)):

        percent = (goal_current[i] / goal_targets[i]) * 100 if goal_targets[i] > 0 else 0
        remaining = goal_targets[i] - goal_current[i]

        months_left = remaining / 100
        if months_left < 0:
            months_left = 0

        print(f"\nGoal: {goal_names[i]}")
        print(f"- Progress: {percent:.1f}%")
        print(f"- Remaining: ${remaining:.2f}")
        print(f"- Time left ($100/mo): {months_left:.1f} months")


        if percent < lowest_percent:
            lowest_percent = percent
            attention_goal = goal_names[i]


        total_target_analysis += goal_targets[i]
        total_current_analysis += goal_current[i]

    overall_progress_analysis = (total_current_analysis / total_target_analysis) * 100 if total_target_analysis > 0 else 0
    total_remaining_analysis = total_target_analysis - total_current_analysis

    print(f"\n" + "="*25)
    print(f"🎯 Overall Goals Summary:")
    print(f"Total target: ${total_target_analysis:.2f}")
    print(f"Total saved: ${total_current_analysis:.2f}")
    print(f"Overall progress: {overall_progress_analysis:.1f}%")
    print(f"Total remaining: ${total_remaining_analysis:.2f}")
    print("="*25)

    print(f"\n💡 ADVICE: Your goal '{attention_goal}' needs the most attention.")
    print(f"It is only {lowest_percent:.1f}% complete.")

    if lowest_percent < 20:
        print("Keep going! You are building a habit that will benefit you in the future don't give up!")
    elif lowest_percent < 50:
        print("You're almost halfway there! Even saving a few cents can help you reach your goals faster, keep track of all the small change!")
    else:
        print("You're doing great stay consistent!")

else:
    print("No goals to analyze!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python
print("\n=== Interactive Finance Menu ===")

# Simple menu system
running = True

while running:
    print("\n" + "-"*30)
    print(f"💰 {name}'s Finance Tracker")
    if len(expenses) > 0:
        print(f"📊 Current: {len(expenses)} expenses, ${sum(expenses):.2f} spent")
    print("-"*30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    if choice == "1":
        # Display expense summary
        print("\n--- Expense Summary ---")

        # Your task: Show all expenses with numbers
        # Format: "1. $23.50 - Coffee"
        # Include total at the end

        if len(expenses) == 0:
            print("No expenses recorded yet!")
        else:
            for i in range(len(expenses)):
                # i+1 makes the list start at 1 instead of 0 for the user
                print(f"{i+1}. ${expenses[i]:.2f} - {descriptions[i]}")
            print(f"Total: ${sum(expenses):.2f}")


    elif choice == "2":
        # Display goals progress
        print("\n--- Goals Progress ---")

        # Your task: Show each goal with progress percentage
        # Format: "Goal Name: $current/$target (XX% complete)"
        if len(goal_names) == 0:
            print("No goals set yet!")
        else:
            for i in range(len(goal_names)):
                # Avoid division by zero if target is 0
                if goal_targets[i] > 0:
                    percent = (goal_current[i] / goal_targets[i]) * 100
                else:
                    percent = 0.0 # If target is 0, percentage is 0
                print(f"{goal_names[i]}: ${goal_current[i]:.2f}/${goal_targets[i]:.2f} ({percent:.1f}% complete)")


    elif choice == "3":
        # Add single new expense
        print("\n--- Add New Expense ---")

        # Your task: Add one more expense to existing lists
        # Use validation pattern, update totals, give feedback
        try:
            new_amount = float(input("Enter expense amount: $"))
            if new_amount > 0:
                new_desc = input("What was this for? ")
                expenses.append(new_amount)
                descriptions.append(new_desc)
                print(f"Success! Added ${new_amount:.2f} for {new_desc}.")
            else:
                print("Amount must be positive!")
        except ValueError:
            print("Invalid input! Please enter a number.")

    elif choice == "4":
        # Budget status check
        print("\n--- Budget Check ---")

        # Your task: Calculate and display budget status
        # Show percentage used, remaining amount, daily budget left
        # Give advice based on budget usage
        total_spent = sum(expenses)
        remaining = monthly_budget - total_spent
        percent_used = (total_spent / monthly_budget) * 100

        print(f"Budget Used: {percent_used:.1f}%")
        print(f"Remaining: ${remaining:.2f}")

        if percent_used > 100:
            print("⚠️ ALERT: You are over budget!")
        elif percent_used > 80:
            print("Status: Caution! You've used most of your budget.")
        else:
            print("Status: You are doing great! Well within budget.")


    elif choice == "5":
        # Exit with summary
        running = False
        print(f"\nGoodbye {name}!")
        total_final = sum(expenses) if expenses else 0
        print(f"Final summary: {len(expenses)} expenses, ${total_final:.2f} total")
        print("🚀 Your finance tracker is growing stronger each week!")

    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```
